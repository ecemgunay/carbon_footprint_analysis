```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Unicode destekleyen font kullan (SUBSCRIPT TWO sorunu çözülür)
plt.rcParams['font.family'] = 'DejaVu Sans'

# Ürün türüne göre karbon salımı
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df,
    x="Greenhouse_Gas_Emissions(kg_CO2e)",
    y="Product_Type",
    estimator=np.mean,
    order=df.groupby("Product_Type")["Greenhouse_Gas_Emissions(kg_CO2e)"]
           .mean()
           .sort_values(ascending=False)
           .index
)
plt.title("Ürün Türüne Göre Ortalama Karbon Salımı (kg CO\u2082e)")
plt.xlabel("Ortalama Karbon Salımı")
plt.ylabel("Ürün Türü")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Marka bazlı karşılaştırma
brand_summary = df.groupby("Company")[[
    "Greenhouse_Gas_Emissions(kg_CO2e)",
    "Water_Consumption(L)",
    "Energy_Consumption(kWh)"
]].mean().round(2)
display(brand_summary)
```

```python
# Strateji maliyeti vs CO2 azaltımı
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Strategy_Cost_USD",
    y="Expected_CO2_Reduction(kg)",
    hue="Company"
)
plt.title("Strateji Maliyeti vs CO₂ Azaltımı")
plt.xlabel("Strateji Maliyeti (USD)")
plt.ylabel("Beklenen CO₂ Azaltımı (kg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Karbon Karnesi hesaplama
def karbon_karnesi(row):
    score = 0
    if row["Greenhouse_Gas_Emissions(kg_CO2e)"] <= df["Greenhouse_Gas_Emissions(kg_CO2e)"].median(): score += 30
    if row["Expected_CO2_Reduction(kg)"] >= 1.5: score += 30
    if row["ROI_Months"] <= 12: score += 20
    if row["Recycled_Material_Use(%)"] >= 30: score += 10
    if row["Renewable_Energy_Use(%)"] >= 50: score += 10
    return score

df["Carbon_Report_Score"] = df.apply(karbon_karnesi, axis=1)

# Skor dağılımı
plt.figure(figsize=(8, 5))
sns.histplot(df["Carbon_Report_Score"], bins=10, kde=True, color="seagreen")
plt.title("Karbon Karnesi Skor Dağılımı")
plt.xlabel("Skor")
plt.ylabel("Ürün Sayısı")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Strateji önerileri
def strateji_onerisi(row):
    öneriler = []
    if row["Greenhouse_Gas_Emissions(kg_CO2e)"] > df["Greenhouse_Gas_Emissions(kg_CO2e)"].median():
        öneriler.append("Malzeme değiştir (geri dönüştürülmüş kullan)")
    if row["Renewable_Energy_Use(%)"] < 40:
        öneriler.append("Yenilenebilir enerji kullanımını artır")
    if row["Energy_Efficiency_Score"] < 70:
        öneriler.append("Enerji verimliliği yatırımı yap")
    if row["Recycled_Material_Use(%)"] < 20:
        öneriler.append("Geri dönüştürülmüş içerik oranını artır")
    if row["ROI_Months"] > 12:
        öneriler.append("Düşük geri dönüşlü stratejilerden kaçın")
    return öneriler[:3]

df["Strategy_Combo"] = df.apply(strateji_onerisi, axis=1)
df[["Company", "Product_Type", "Strategy_Combo", "Carbon_Report_Score"]].head()
```

```python
# Risk ve Fırsat Analizi

# Regülasyon Riski Dağılımı Grafiği
plt.figure(figsize=(6, 4))
sns.countplot(x="Regulation_Risk", hue="Regulation_Risk", data=df, palette="pastel", legend=False)
plt.title("Regülasyon Riski Dağılımı")
plt.xlabel("Risk Seviyesi")
plt.ylabel("Ürün Sayısı")
plt.tight_layout()
plt.grid(True)
plt.show()

# Tüketici Eğilimi vs Karbon Salımı
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x="Consumer_Trend_Score",
    y="Greenhouse_Gas_Emissions(kg_CO2e)",
    hue="Regulation_Risk",
    data=df
)
plt.title("Tüketici Eğilimi vs Karbon Salımı (Regülasyon Riskine Göre)")
plt.xlabel("Tüketici Eğilim Skoru")
plt.ylabel("Karbon Salımı (kg CO₂e)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Otomatik Fırsat Yorumu Oluştur
def fırsat_değerlendirme(row):
    if row["Consumer_Trend_Score"] >= 80 and row["Greenhouse_Gas_Emissions(kg_CO2e)"] <= 3.0:
        return "Yüksek tüketici ilgisiyle öne çıkarılabilir"
    elif row["Consumer_Trend_Score"] < 70 and row["Greenhouse_Gas_Emissions(kg_CO2e)"] > 4.5:
        return "İmaj riski yüksek, dönüşüm önerilir"
    else:
        return "Dengeli"

df["Opportunity_Insight"] = df.apply(fırsat_değerlendirme, axis=1)

# Örnek Tablo Göster
df[[
    "Company", "Product_Type", "Greenhouse_Gas_Emissions(kg_CO2e)",
    "Consumer_Trend_Score", "Regulation_Risk", "Opportunity_Insight"
]].head()
```
