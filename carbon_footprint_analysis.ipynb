```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Unicode destekleyen font kullan (SUBSCRIPT TWO sorunu Ã§Ã¶zÃ¼lÃ¼r)
plt.rcParams['font.family'] = 'DejaVu Sans'

# ÃœrÃ¼n tÃ¼rÃ¼ne gÃ¶re karbon salÄ±mÄ±
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df,
    x="Greenhouse_Gas_Emissions(kg_CO2e)",
    y="Product_Type",
    estimator=np.mean,
    order=df.groupby("Product_Type")["Greenhouse_Gas_Emissions(kg_CO2e)"]
           .mean()
           .sort_values(ascending=False)
           .index
)
plt.title("ÃœrÃ¼n TÃ¼rÃ¼ne GÃ¶re Ortalama Karbon SalÄ±mÄ± (kg CO\u2082e)")
plt.xlabel("Ortalama Karbon SalÄ±mÄ±")
plt.ylabel("ÃœrÃ¼n TÃ¼rÃ¼")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Strateji maliyeti vs CO2 azaltÄ±mÄ±
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Strategy_Cost_USD",
    y="Expected_CO2_Reduction(kg)",
    hue="Company"
)
plt.title("Strateji Maliyeti vs COâ‚‚ AzaltÄ±mÄ±")
plt.xlabel("Strateji Maliyeti (USD)")
plt.ylabel("Beklenen COâ‚‚ AzaltÄ±mÄ± (kg)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Karbon Karnesi hesaplama
def karbon_karnesi(row):
    score = 0
    if row["Greenhouse_Gas_Emissions(kg_CO2e)"] <= df["Greenhouse_Gas_Emissions(kg_CO2e)"].median(): score += 30
    if row["Expected_CO2_Reduction(kg)"] >= 1.5: score += 30
    if row["ROI_Months"] <= 12: score += 20
    if row["Recycled_Material_Use(%)"] >= 30: score += 10
    if row["Renewable_Energy_Use(%)"] >= 50: score += 10
    return score

df["Carbon_Report_Score"] = df.apply(karbon_karnesi, axis=1)

# Skor daÄŸÄ±lÄ±mÄ±
plt.figure(figsize=(8, 5))
sns.histplot(df["Carbon_Report_Score"], bins=10, kde=True, color="seagreen")
plt.title("Karbon Karnesi Skor DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Skor")
plt.ylabel("ÃœrÃ¼n SayÄ±sÄ±")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Strateji Ã¶nerileri
def strateji_onerisi(row):
    Ã¶neriler = []
    if row["Greenhouse_Gas_Emissions(kg_CO2e)"] > df["Greenhouse_Gas_Emissions(kg_CO2e)"].median():
        Ã¶neriler.append("Malzeme deÄŸiÅŸtir (geri dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lmÃ¼ÅŸ kullan)")
    if row["Renewable_Energy_Use(%)"] < 40:
        Ã¶neriler.append("Yenilenebilir enerji kullanÄ±mÄ±nÄ± artÄ±r")
    if row["Energy_Efficiency_Score"] < 70:
        Ã¶neriler.append("Enerji verimliliÄŸi yatÄ±rÄ±mÄ± yap")
    if row["Recycled_Material_Use(%)"] < 20:
        Ã¶neriler.append("Geri dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lmÃ¼ÅŸ iÃ§erik oranÄ±nÄ± artÄ±r")
    if row["ROI_Months"] > 12:
        Ã¶neriler.append("DÃ¼ÅŸÃ¼k geri dÃ¶nÃ¼ÅŸlÃ¼ stratejilerden kaÃ§Ä±n")
    return Ã¶neriler[:3]

df["Strategy_Combo"] = df.apply(strateji_onerisi, axis=1)
df[["Company", "Product_Type", "Strategy_Combo", "Carbon_Report_Score"]].head()
```

```python
# Risk ve FÄ±rsat Analizi

# RegÃ¼lasyon Riski DaÄŸÄ±lÄ±mÄ± GrafiÄŸi
plt.figure(figsize=(6, 4))
sns.countplot(x="Regulation_Risk", hue="Regulation_Risk", data=df, palette="pastel", legend=False)
plt.title("RegÃ¼lasyon Riski DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Risk Seviyesi")
plt.ylabel("ÃœrÃ¼n SayÄ±sÄ±")
plt.tight_layout()
plt.grid(True)
plt.show()

# TÃ¼ketici EÄŸilimi vs Karbon SalÄ±mÄ±
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x="Consumer_Trend_Score",
    y="Greenhouse_Gas_Emissions(kg_CO2e)",
    hue="Regulation_Risk",
    data=df
)
plt.title("TÃ¼ketici EÄŸilimi vs Karbon SalÄ±mÄ± (RegÃ¼lasyon Riskine GÃ¶re)")
plt.xlabel("TÃ¼ketici EÄŸilim Skoru")
plt.ylabel("Karbon SalÄ±mÄ± (kg COâ‚‚e)")
plt.grid(True)
plt.tight_layout()
plt.show()

# Otomatik FÄ±rsat Yorumu OluÅŸtur
def fÄ±rsat_deÄŸerlendirme(row):
    if row["Consumer_Trend_Score"] >= 80 and row["Greenhouse_Gas_Emissions(kg_CO2e)"] <= 3.0:
        return "YÃ¼ksek tÃ¼ketici ilgisiyle Ã¶ne Ã§Ä±karÄ±labilir"
    elif row["Consumer_Trend_Score"] < 70 and row["Greenhouse_Gas_Emissions(kg_CO2e)"] > 4.5:
        return "Ä°maj riski yÃ¼ksek, dÃ¶nÃ¼ÅŸÃ¼m Ã¶nerilir"
    else:
        return "Dengeli"

df["Opportunity_Insight"] = df.apply(fÄ±rsat_deÄŸerlendirme, axis=1)

# Ã–rnek Tablo GÃ¶ster
df[[
    "Company", "Product_Type", "Greenhouse_Gas_Emissions(kg_CO2e)",
    "Consumer_Trend_Score", "Regulation_Risk", "Opportunity_Insight"
]].head()
```
